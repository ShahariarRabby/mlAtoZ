```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
clf = RandomForestRegressor(n_estimators=10, random_state=0)
```

```python
clf.fit(X,y)
```

```python
pred = clf.predict(X)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
X_grid = np.arange(min(X), max(X), 0.0001) # choice of 0.01 instead of 0.1 step because the data is feature scaled
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
# plt.plot(X_grid, clf.predict(X_grid), color = 'blue')
plt.scatter(X, clf.predict(X), color = 'g')
plt.title('Truth or Bluff (SVR)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
