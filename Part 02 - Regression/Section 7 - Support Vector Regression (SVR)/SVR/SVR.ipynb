```python
import numpy as np
import pandas as pd
from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,1:2]
y = dataset['Salary']
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
```

```python
scl = StandardScaler()
```

```python
X = scl.fit_transform(X)
```

```python
y = scl.fit_transform(y)
```

```python
clf = SVR(kernel='rbf')
```

```python
clf.fit(X,y)
```

```python
pred = clf.predict(X)
```

```python
plt.plot(X,pred)
plt.scatter(X,y)
plt.grid
```
